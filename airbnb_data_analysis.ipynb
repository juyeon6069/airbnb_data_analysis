```python
# Load the file
import pandas as pd

data = pd.read_csv('AB_NYC_2019.csv')
```

<Question 1>     
    I did 3 things for cleaning data. First, I deleted useless data. I think last_review is unnecessary data because most of the information was represented by ###, which is meaningless. Second, I deleted duplicated data. If latitude and longitude are the same, I thought they were the same data and deleted them. Third, I replaced empty cells. If empty cells in name and host_name were replaced with empty strings. The empty cell in review_per_month was replaced by 0, and the empty cell in availability_365 was replaced by '?'.

```python
# Question1: Cleaning Data

# 1. Delete Useless Data - last_review
data = data.drop('last_review', axis=1)

# 2. Delete Duplicated Data - By checking latitude and longitude
duplicated_rows = data.duplicated(subset=['latitude', 'longitude'])
data.drop_duplicates(inplace=True)

# 3. Replace Empty Cell - name / host_name / review_per_month / availability_365
empty_columns = data.columns[data.isnull().any()]
    
# 3-1. name -> replace to empty string
data['name'].fillna('', inplace=True)

# 3-2. host_name -> replace to empty string
data['host_name'].fillna('', inplace=True)

# 3-3: review_per_month -> replace to 0
data['reviews_per_month'].fillna(0, inplace=True)

# 3-4: availability_365 -> replace to string, '?'
data['availability_365'].fillna('?', inplace=True)
```

<Question 2>    
       a. To find the top five neighborhoods, I first grouped the data with the same neighborhood and filtered the regions with more than five lists. After that, use the .nlargest(), .nsmallest() functions to print out the five neighborhood with the largest or smallest average price among the groups.   
       b. If looking at the size of the box of the plot, the higher the price, the higher the height of the box. Therefore, we can know that the higher the price, the higher the price fluctuation. 

```python
# Question2

# a.
# filtering them with 5 or more lists
filtered_neighbourhood = data.groupby('neighbourhood').filter(lambda x: len(x) > 5)

# group by neighbourhood and measure its mean of the price
grouped_neighbourhood = filtered_neighbourhood.groupby('neighbourhood')['price'].mean()

#find top 5 price and bottom 5 price
top_5 = grouped_neighbourhood.nlargest(5)
bottom_5 = grouped_neighbourhood.nsmallest(5)

print("Top 5 neighborhoods based on the price:")
print(top_5)
print()

print("Bottom 5 neighborhoods based on the price:")
print(bottom_5)

# b.
import matplotlib.pyplot as plt

data.boxplot(column='price',by='neighbourhood_group', showfliers=False)
plt.suptitle(None)
plt.grid(False)

plt.title("Price Variation between Different Neighborhood Groups")
plt.xlabel("neighbourhood group")
plt.ylabel("price")

plt.show()
```

<Question 3>     
    Interesting features were price, number_of_reviews, reviews_per_month, and availability_365. Because I was curious about how much price, the number of reviews, and availability are related. For this, calculated the correlation of four columns and expressed it as a heatmap. As a result, availability_365 was more related to the price than to the number of reviews. And of course, the total number of reviews and the number of reviews per month showed a large correlation. 
    In addition,  to find the most positive and most negative correlation, the values were sorted in order of correlation and found as the first one and last one.

```python
import seaborn as sns

# Question 3

# pick interesting data
interesting_data_list = ['price', 'number_of_reviews', 'reviews_per_month', 'availability_365']
interesting_data = data[interesting_data_list]

# find Pearson correlation analysis
pearson = interesting_data.corr()

# represented by heat map
sns.heatmap(pearson, annot=True)

plt.title("Pairwise Pearson Correlation Analysis")
plt.show()

# find most positive and most negative correlations
pearson_values = pearson.values

most_positive = pearson.unstack().sort_values(ascending=False)[len(interesting_data_list)]
print('most positive correlation: ', most_positive)
print()

most_negative = pearson.unstack().nsmallest(1)[0]
print('most negative correlation: ',most_negative)
```

<Question 8-1>      
    First plot that I create is line graph. Because I think the relationship between price and the number of reviews is interesting. Before I saw the results, I thought that if it was expensive, it would be more likely to be a good airbnb. So I expected that there would be a lot of reviews. However, on the contrary, the more expensive the results were, the less reviews there were. There may be many reasons, but I thought that the reviews might have been bad or the reason why the price was high might be because of the lack of supply, not because of the good airbnb.

```python
import numpy as np

neighbourhood_stats = data.groupby('neighbourhood_group').agg({'price': 'mean', 'reviews_per_month': 'mean'}).reset_index().sort_values(by='price', ascending=False)

plt.figure(figsize=(10, 6))

bar_width = 0.35
X_axis = np.arange(len(neighbourhood_stats['neighbourhood_group'])) 

plt.plot(X_axis, neighbourhood_stats['price'], marker='o', label='Mean Price', color='skyblue')
plt.plot(X_axis, neighbourhood_stats['reviews_per_month']*100, marker='o', label='Mean Reviews per Month *100', color='orange')
plt.xlabel('Neighborhood Group')
plt.ylabel('Mean Value')
plt.title('Mean Price and Mean Reviews per Month by Neighborhood')
plt.xticks(X_axis, neighbourhood_stats['neighbourhood_group'])
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
